```python
# Import relevant libraries and packages.
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns # For all our visualization needs.
import statsmodels.api as sm 
from statsmodels.graphics.api import abline_plot  
from sklearn.metrics import mean_squared_error, r2_score 
from sklearn.model_selection import train_test_split 
from sklearn import linear_model, preprocessing  
import warnings # For handling error messages.
from scipy.stats import normaltest
```

```python
# import cleaned credit card "train" dataset csv     cc_classify_out_test.csv    cc_classify_out_train.csv   add  index_col=False
cc_classify_train_EDA = pd.read_csv('C:\\Users\\\Robert\\.spyder-py3\\raw_data\\cc_classify_out_train.csv', index_col=False)
#
df = cc_classify_train_EDA
```

```python
# view imported dataset
df.head(10)
```

```python
# drop columns:  Customer_ID, Name, SSN, Age ,  ID, MonthCredit_Score, Payment_Behaviour, Type_of_Loan
df.drop([ 'Customer_ID', 'Name', 'SSN', 'Age', 'ID', 'Month', 'Credit_Score', 'Payment_Behaviour', 'Type_of_Loan'], inplace=True, axis=1)
```

```python
# rename column name fn_Age to Age
df.rename(columns={'fn_Age': 'Age'}, inplace=True)
# rename column name fn_IRate to Interest_Rate
df.rename(columns={'fn_IRate': 'Interest_Rate'}, inplace=True)
# #rename column name fn_Credit_History_Age to Credit_History_Age
df.rename(columns={'fn_Credit_History_Age': 'Credit_History_Age'}, inplace=True)
```

```python
#df.describe
df['Occupation'] = df['Occupation'].str.replace('_______', 'None_Listed')
df['Occupation'].unique()
```

```python
# perform one-hot encoding using get_dummies()
one_hot = pd.get_dummies(df['Occupation'])

# concatenate the original DataFrame with the one-hot encoded columns
df = pd.concat([df, one_hot], axis=1)
#
df.drop([ 'Occupation'], inplace=True, axis=1)
#
print(df.head())
```

```python
df.drop([ 'Unnamed: 0'], inplace=True, axis=1)
print(df.head())
```

```python
df['Credit_Mix'].unique()
# perform one-hot encoding using get_dummies()
one_hot = pd.get_dummies(df['Credit_Mix'], prefix='Credit_Mix')

# concatenate the original DataFrame with the one-hot encoded columns
df = pd.concat([df, one_hot], axis=1)
#
df.drop([ 'Credit_Mix'], inplace=True, axis=1)
```

```python
df['Payment_of_Min_Amount'].unique()

# perform one-hot encoding using get_dummies()
one_hot = pd.get_dummies(df['Payment_of_Min_Amount'],prefix='PMT_MinAmt')

# concatenate the original DataFrame with the one-hot encoded columns
df = pd.concat([df, one_hot], axis=1)
#
df.drop([ 'Payment_of_Min_Amount'], inplace=True, axis=1)
```
